```python
# Импортирование необходимых библиотек
import numpy as np  # Для работы с числовыми данными и массивами
import matplotlib.pyplot as plt  # Для создания графиков и визуализации данных
import skfuzzy as fuzz  # Для работы с нечеткой логикой и функциями принадлежности
from matplotlib.widgets import Cursor  # Для добавления интерактивного курсора на график

# Параметры для лингвистической переменной
ling_var_name = "Температура печи"  # Имя лингвистической переменной
num_terms = 5  # Количество термов, описывающих лингвистическую переменную
term_names = ["Очень низкая", "Низкая", "Оптимальная", "Высокая", "Очень высокая"]  # Имена термов
max_scale_value = 1500  # Максимальное значение шкалы (максимальная температура печи в градусах Цельсия)

# Точки шкалы, где функции принадлежности принимают значения 0 и 1
zero_one_points = [
    (None, None, 800, 1000),  # Для терма "Очень низкая"
    (800, 1000, 1200, 1300),  # Для терма "Низкая"
    (1200, 1300, 1350, 1400),  # Для терма "Оптимальная"
    (1350, 1400, 1450, 1500),  # Для терма "Высокая"
    (1450, 1500, None, None)  # Для терма "Очень высокая"
]

# Создание линейного пространства для оси x от 700 до max_scale_value
x = np.linspace(start=700, stop=max_scale_value, num=(num_terms * 100))

# Создание графика
fig, ax = plt.subplots(figsize=(10, 5))
plt.title(ling_var_name)
plt.ylabel("Принадлежность")
plt.xlabel("Температура (°C)")
plt.grid(True)

# Создание функций принадлежности и добавление их на график
membership_functions = []
for term_name, points in zip(term_names, zero_one_points):
    # Заменяем None на крайние значения шкалы
    points = [
        0 if point is None and i < 2 else max_scale_value if point is None else point
        for i, point in enumerate(points)
    ]
    membership_function = fuzz.trapmf(x, points)
    membership_functions.append((term_name, membership_function))
    plt.plot(x, membership_function, label=term_name)

# Обновление легенды графика
plt.legend()

# Добавление курсора для интерактивности
cursor = Cursor(ax, useblit=True, color='red', linewidth=1)

# Функция для обработки кликов мыши
def onclick(event):
    if event.inaxes:
        temp_value = event.xdata  # Получение значения температуры (ось x)
        membership_values = []
        for term_name, mf in membership_functions:
            membership_value = fuzz.interp_membership(x, mf, temp_value)
            membership_values.append((term_name, membership_value))

        print(f"\nТемпература: {temp_value:.2f} °C")
        for term_name, value in membership_values:
            print(f"Принадлежность к '{term_name}': {value:.2f}")

# Привязка обработчика кликов мыши к функции onclick
fig.canvas.mpl_connect('button_press_event', onclick)

# Отображение графика
plt.show()
```

```python
import numpy as np  # Для работы с числовыми данными и массивами
import matplotlib.pyplot as plt  # Для создания графиков и визуализации данных
import skfuzzy as fuzz  # Для работы с нечеткой логикой и функциями принадлежности
from matplotlib.widgets import Cursor  # Для добавления интерактивного курсора на график

# Параметры для лингвистической переменной
ling_var_name = "Температура печи"
num_terms = 5
term_names = ["Очень низкая", "Низкая", "Оптимальная", "Высокая", "Очень высокая"]
max_scale_value = 1500

# Расширенные точки шкалы для функций принадлежности (вариант 11)
zero_one_points = [
    (None, None, 800, 1000),      # Очень низкая
    (800, 1000, 1150, 1250),     # Низкая
    (1150, 1250, 1350, 1450),    # Оптимальная
    (1350, 1450, 1500, None),    # Высокая
    (None, None, None, None)     # Очень высокая — сделаем константу 1
]

# Создаем линейное пространство по оси X (температура от 700 до 1500)
x = np.linspace(start=700, stop=max_scale_value, num=num_terms * 100)

fig, ax = plt.subplots(figsize=(10, 5))
plt.title(ling_var_name)
plt.ylabel("Принадлежность")
plt.xlabel("Температура (°C)")
plt.grid(True)

membership_functions = []

for term_name, points in zip(term_names, zero_one_points):
    # Обработка None для крайних значений шкалы
    # Если все точки None — значит функция принадлежности равна 1 на всем диапазоне
    if all(p is None for p in points):
        membership_function = np.ones_like(x)
    else:
        points = [0 if (point is None and i < 2) else
                  max_scale_value if (point is None and i >= 2) else point
                  for i, point in enumerate(points)]
        membership_function = fuzz.trapmf(x, points)
    membership_functions.append((term_name, membership_function))
    plt.plot(x, membership_function, label=term_name)

plt.legend()
cursor = Cursor(ax, useblit=True, color='red', linewidth=1)

# Аннотация для интерактивного отображения принадлежности
annotation = ax.annotate(
    '', xy=(0, 0), xytext=(20, 20), textcoords='offset points',
    bbox=dict(boxstyle="round,pad=0.3", edgecolor="black", facecolor="white"),
    arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=0")
)
annotation.set_visible(False)

def update_annotation(event, membership_values):
    temp_value = event.xdata
    text = f"Температура: {temp_value:.2f} °C\n"
    for term_name, value in membership_values:
        text += f"{term_name}: {value:.2f}\n"
    annotation.xy = (event.xdata, event.ydata)
    annotation.set_text(text)
    annotation.set_visible(True)
    fig.canvas.draw_idle()

def onclick(event):
    if event.inaxes:
        temp_value = event.xdata
        membership_values = []
        for term_name, mf in membership_functions:
            membership_value = fuzz.interp_membership(x, mf, temp_value)
            membership_values.append((term_name, membership_value))
        update_annotation(event, membership_values)

fig.canvas.mpl_connect('button_press_event', onclick)

plt.show()
```
